```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'              #Window 용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False


x = [1,2,3]
y = [2,4,8]

plt.xlabel("x line",color = 'blue', loc = 'center')
plt.ylabel("y line",color = '#8c4e37', loc = 'bottom')

# 눈금 표시
plt.xticks([1,2,3])
plt.yticks([2,4])


plt.plot(x,y)
plt.show()



```
