```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'              #Window 용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False

df= pd.read_excel('score.xlsx',index_col='지원번호')


fig,axs = plt.subplots(2,2,figsize = (15,13))
fig.suptitle('4개 복합 그래프')


axs[0,0].bar(df['이름'],df['국어'],label = '국어')
axs[0,0].set_title('1.국어점수 그래프')
axs[0,0].set(xlabel = '이름', ylabel = '점수')
axs[0,0].set_facecolor('lightyellow')
axs[0,0].legend(loc='upper center')


axs[0,1].set_title('2. 영어,수학점수 그래프')
axs[0,1].plot(df['이름'],df['영어'],label = '영어')
axs[0,1].plot(df['이름'],df['수학'],label = '수학')
axs[0,1].legend()


axs[1,0].set_title('3. 가로바 키 그래프')
axs[1,0].set_xlim(100,230)
axs[1,0].barh(df['이름'],df['키'])
for i in range(8):
  axs[1,0].text(df['키'][i]+5,df['이름'][i],df['키'][i],ha = 'center')


axs[1,1].set_title('4. 사회점수 그래프')
axs[1,1].plot(df['이름'],df['사회'], color='g', alpha=0.5)
for i in range(8):
  axs[1,1].text(df['이름'][i],df['사회'][i]+5,df['사회'][i],ha = 'center')


plt.show()



df
```
