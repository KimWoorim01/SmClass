```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

matplotlib.rcParams['font.family'] = 'Malgun Gothic'              #Window 용
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False


df = pd.read_excel('score.xlsx',index_col='지원번호')


values = [30,25,20,13,10,2]
explode = [0.3,0.1,0,0,0,0]


labels = ['python', 'java', 'javascript', 'oracle','html','etc']
plt.title('profram ')

wedgeprops = {'width':0.6, 'edgecolor': 'red'}


plt.pie(values,labels=labels, explode=explode, autopct = '%.1f%%',counterclock=False, startangle=90, wedgeprops=wedgeprops, pctdistance=0.7)


plt.legend(loc=(1.2,0.3) ,title = '언어별 선호도')

plt.show()
```

```python
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']

## 원그래프
# values = [30,25,20,13,10,2]  # 100퍼센트
values = [1,1,1,1,2,1]  # 100퍼센트
explode = [0.2,0.1,0,0,0,0]
labels = ['python','java','javascript','oracle','html','etc']
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']
plt.title('프로그램 사용률')
## pie함수(데이터값), labels : 글자출력
## explode : 원그래프 분리, countclock : 시계반대방향, False : 시계방향
## startangle = 90 : 12시부터 시작됨.
## 데이터 값 출력 : autopct
## colors : 색상변경
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90,colors=colors)
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
# plt.legend(loc='upper right')
plt.show()
```
