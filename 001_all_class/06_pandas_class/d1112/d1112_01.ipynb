```python
import pandas as pd
import random


def weekFunc(x):

  if x == 0: return "일요일"
  elif x == 1: return "월요일"
  elif x == 2: return "화요일"
  elif x == 3: return "수요일"
  elif x == 4: return "목요일"
  elif x == 5: return "금요일"
  elif x == 6: return "토요일"

df = pd.read_excel('score.xlsx', index_col='지원번호')

df['요일'] = 1

df.loc['1번','요일'] = 5

df.shape

for i in range(8):
  df.iloc[i,9] = random.randint(0,6)

df['요일'] = df['요일'].apply(weekFunc)
# df

df['결과'] = '불합격'

filt = (df['국어']>=60)

df.loc[filt,'결과'] = '합격'

df['키'] = df['키'].astype(str) + "cm"

df

df.groupby("학교")["이름"].count()
df.groupby("학교")["국어"].mean()
df.groupby("학교")["이름"].sum()

df['학년'] = [3,2,2,1,1,2,1,1]
df

df.groupby(['학교'])['국어'].agg(['mean','min','max','sum'])
df.groupby(['학교','학년'])['국어'].agg(['mean','min','max','sum'])


# mean(), max(), min(), std()           #컬럼이 숫자가 아니면 동작안함

```
